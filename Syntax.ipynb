```python
#data annotation with UDpipe
def make_conll_with_udpipe(text):
    model_path = ''
    model = Model.load(model_path)
    pipeline = Pipeline(model, 'tokenize', Pipeline.DEFAULT, Pipeline.DEFAULT, 'conllu')
    return pipeline.process(text)

#Path to the directory where texts to preprocess are
path = ''
#Path to the directory where preprocessed texts should be saved
savepath = ''
for file in os.listdir(path):
    with open(path+file, encoding='UTF-8') as txt:
        text = txt.read()
        x = make_conll_with_udpipe(text)
        filename = str(file)
    with open(savepath+filename, 'w', encoding='UTF-8') as newfile:
        newfile.write(x)
```

```python
#Main clause number (just in case; that is also easy to get by other ways)
path = ''
rootclauses=[]
for file in os.listdir(path):
    conll = pyconll.load_from_file(path+file)
    cases=[]
    for sentence in conll:
        for word in sentence:
            if word.deprel=='root':
                cases.append(word.form)
    rootclauses.append(len(cases))
```
